```python
collectPerms = []

# both parameters are array of strings
def permute(chooseFrom, chosen):
    if len(chooseFrom) == 0:
        collectPerms.append("".join(chosen))
        return
    
    for i in range(len(chooseFrom)):
        
        char = chooseFrom[i]
        chooseFrom[i] = chooseFrom[-1]
        chooseFrom.pop()
        
        chosen += char
        
        # recurse
        permute(chooseFrom, chosen)
        
        #undo
        chooseFrom.append(char)
        chooseFrom[i], chooseFrom[-1] = chooseFrom[-1], chooseFrom[i]
        chosen.pop()
```

```python
chooseFrom = [i for i in "Andres"]
chosen = []
permute(chooseFrom, chosen)
```
